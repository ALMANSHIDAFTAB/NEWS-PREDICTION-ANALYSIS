# NOW WE WILL ADD THE COLUMNS IN X AND Y VARIABLE FOR ML

```python
x=np.array(df['title'])
y=np.array(df['category'])
```

# The CountVectorizer provides a simple way to both tokenize a collection of text documents and build a vocabulary of known words, but also to encode new documents using that vocabulary

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
a=CountVectorizer()
```

# NOW WE WILL FIT AND TRANSFORM THE DATA FOR ML OPERATION

```python
ax=a.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

# SPLITTING THE DATA IN FOR TESTING AND TRAINING 

```python
xTrain, xTest, yTrain, yTest = train_test_split(ax, y, test_size=0.1, random_state=2)
```

## FOR  HEIGHER ACCURACY WE TOOK 10%DATA FOR TEST AND REST IS TRAINED

```python
from sklearn.naive_bayes import MultinomialNB
```

# Naive Bayes classifier for multinomial model,The multinomial Naive Bayes classifier is suitable for classification with discrete features (e.g., word counts for text classification).

```python
b=MultinomialNB()
```

# FITTING THE DATA INTO MULTINOMIALNB

```python
b.fit(xTrain,yTrain)
```

## ADDING A TEXTFIELD TIO CHECK THE WORK OR SENTENCE AND ITS NEWS TYPE


```python
z=input("enter the text:  ")
```

# TRANSFORMING THE TEXT INTO ARRAY FOR PREDICTION

```python
msg=a.transform([z]).toarray()
```

```python
pre=b.predict(msg)
```

# PRINTING THE NEWS TYPE(YOU CAN ALSO CHECK IT)

```python
from sklearn.metrics import accuracy_score
```

# TESTING THE ACCURACY!

```python
accuracy_score(yTest,b.predict(xTest))
```
